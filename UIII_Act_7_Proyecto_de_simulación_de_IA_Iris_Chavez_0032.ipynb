```python
# --- SIMULADOR DE IA: MAESTRO PLOMERO Y APRENDIZ ---

print("SISTEMA DE ENTRENAMIENTO VIRTUAL 'PLOMERO-INTELLIGENCE v2.0'")
print("Bienvenido, Aprendiz. Estoy aquí para guiarte en la obra.")

sesion_activa = "si"

# El ciclo while mantiene la IA alerta hasta que el aprendiz termine la jornada
while sesion_activa == "si":
    print("\n*********************************")
    print("MENÚ DE COMANDOS DEL APRENDIZ:")
    print("1. Solicitar herramienta para un trabajo")
    print("2. Reportar una avería para diagnóstico")
    print("3. Finalizar jornada")
    print("*********************************")

    comando = input("¿Qué necesitas hacer? (1/2/3): ")

    # Estructura if-elif-else para procesar la opción del menú
    if comando == "1":
        print("\n--- ASISTENTE DE HERRAMIENTAS ---")
        tipo_trabajo = input("¿El trabajo es 'instalar', 'destapar' o 'ajustar'? ").lower()

        # La IA decide la herramienta usando lógica relacional
        if tipo_trabajo == "instalar":
            print("IA Maestro: Necesitarás la tubería de cobre y el soplete.")
        elif tipo_trabajo == "destapar":
            print("IA Maestro: Trae la sonda mecánica y el líquido desatascador.")
        elif tipo_trabajo == "ajustar":
            print("IA Maestro: La llave inglesa es suficiente para este trabajo.")
        else:
            print("IA Maestro: Esa tarea no está en mi base de datos de plomería.")

    elif comando == "2":
        print("\n--- DIAGNÓSTICO DE AVERÍAS ---")
        gravedad = input("¿Cuál es el nivel de urgencia de la fuga? (bajo/medio/alto): ").lower()

        # Simulación de razonamiento condicional complejo
        if gravedad == "alto":
            print("IA Maestro: ¡CRÍTICO! Corta el suministro general de agua de inmediato.")
        elif gravedad == "medio":
            print("IA Maestro: Coloca cinta de teflón y aprieta las conexiones.")
        else:
            print("IA Maestro: Es solo un goteo. Revisa el empaque cuando termines lo demás.")

    elif comando == "3":
        print("\nIA Maestro: Guardando progreso del aprendizaje... ¡Buen trabajo hoy!")
        # Cambiamos la variable para romper la condición del ciclo while
        sesion_activa = "no"

    else:
        # Manejo de opciones fuera del rango del menú
        print("\n[Error de sistema]: Comando no reconocido. Intenta de nuevo.")

print("\n--- PROGRAMA FINALIZADO ---")
print("Programa realizado por Iris Chavez NL = 11")
```
